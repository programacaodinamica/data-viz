```python
df_covid = pd.read_csv("/content/drive/MyDrive/Colab Datasets/HIST_PAINEL_COVIDBR_15jan2021.csv", sep=';')
```

```python
df_covid.head()
```

```python
covid_estados = df_covid.dropna( subset=['estado'])
```

```python
covid_estados = covid_estados[pd.isnull(covid_estados['municipio'])]
```

```python
covid_estados['letalidade'] = (covid_estados['obitosAcumulado']/covid_estados['casosAcumulado']).fillna(0)
```

```python
covid_estados.head()
```

```python
covid_estados.to_csv('covid_estados_15jan21.csv')
```

```python
df_covid_estados_bcr = pd.DataFrame(covid_estados.groupby(["estado","regiao","data"]).sum()['obitosAcumulado']).unstack()
```

```python
df_covid_estados_bcr.columns = df_covid_estados_bcr.columns.droplevel()
```
