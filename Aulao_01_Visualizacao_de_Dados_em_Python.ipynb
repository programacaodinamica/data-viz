```
municipios = pd.read_csv('/content/drive/My Drive/Colab Datasets/MC01/brasil-municipios.csv')
idhm = pd.read_csv('/content/drive/My Drive/Colab Datasets/MC01/atlas2013_dadosbrutos_idhm_pt.csv', sep=';', decimal=',')
covid = pd.read_csv('/content/drive/My Drive/Colab Datasets/MC01/HIST_PAINEL_COVIDBR_04nov2020.csv', sep=';')
```

```
covid_estados = covid.dropna( subset=['estado'])
```

```
covid_estados.head()
```

```
covid_estados.tail()
```

```
covid_estados.info()
```

```
covid_estados = covid_estados[pd.isnull(covid_estados['municipio'])]
```

```
covid_estados.info()
```

```
covid_estados
```

```
covid_estados['data_format'] = pd.to_datetime(covid_estados['data']).dt.strftime('%d-%m')
```

```
capitais_estados_br = municipios[municipios['capital']==1]
```

```
capitais_estados_br
```

```
municipios.drop_duplicates(subset=['codigo_uf'])
```

```
covid_estados_georref = covid_estados.merge(capitais_estados_br,left_on='coduf', right_on='codigo_uf')
```

```
covid_estados_georref
```

```
import plotly.express as px
```

```
fig = px.scatter_mapbox(covid_estados_georref, lat="latitude", lon="longitude", size="obitosAcumulado",
                 size_max=45, zoom=3, 
                 center={"lat": -13.702797, "lon": -69.6865109},
                 animation_frame='data_format', hover_name='estado', height=800,
                 mapbox_style="carto-positron")
fig.update_layout(
 title='Espalhamento de COVID-19 nos Estados Brasileiros de 25/02 a 04/11')
fig.show()
```
