```
covid_municipios = covid.merge(idhm,left_on='codmun', right_on='Codmun6').merge(municipios,left_on='Codmun7',right_on='codigo_ibge')
```

```
covid_municipios.head()
```

```
covid_municipios.sort_values(by='data')
```

```
import plotly.express as px
```

```
fig = px.scatter_mapbox(covid_municipios[covid_municipios['estado']=='RJ'], lat="latitude", lon="longitude", size="obitosAcumulado",
                 size_max=45, zoom=3, color='IDHM',
                 center={"lat": -13.702797, "lon": -69.6865109},
                 animation_frame='data_format', hover_name='municipio', width=1200, height=800,
                 mapbox_style="carto-positron")
fig.show()
```

```
covid_municipios[covid_municipios.data==pd.Timestamp('2020-07-24')].describe()
```

```
fig_scatter = px.scatter(covid_municipios[covid_municipios.data==pd.Timestamp('2020-07-24')], x="casosAcumulado", y="letalidade", size="obitosAcumulado",
           hover_name="municipio", size_max=60, facet_col="regiao",  color="IDHM",
            range_y=[-0.01,0.2], range_x=[-10000,300000])
fig_scatter.show()
```
